# Numpy ndarray

### [ ndarray , (shape),  ]

```python
import numpy as np
```

```python
A1 = np.array([1,2,3])
print('A1 type: ', type(A1))
print('A1 array shape: ', A1.shape)

A2 = np.array([[1,2,3], 
               [4,5,6]])
print('A2 type: ', type(A2))
print('A2 array shape: ', A2.shape)

A3 = np.array([[1,2,3]])
print('A3 type: ', type(A3))
print('A3 array shape: ', A3.shape)
```

### [ ndarray list type  ]

```python
L1 = [1,2,3]
print('L1: ', L1)
print('L1 type: ', type(L1))

A1 = np.array(L1)
print('A1: ', A1)
print('A1 type: ', type(A1))
```

### [ ndarray     ndarray.dtype ]

```python
L1 = [1,2,3]
print(type(L1))

A1 = np.array(L1)
print(type(A1))
print(A1, A1.dtype)
```

### [ ndarray      ]

```python
L2 = [1, 2, 'test'] #  -> 
A2 = np.array(L2)
print(A2, A2.dtype)

L3 = [1, 2, 3.0]    #  -> 
A3 = np.array(L3)
print(A3, A3.dtype)
```

### [ ndarray     ndarray.astype() ]

```python
A_int1 = np.array([1,2,3])
A_float1 = A_int1.astype('float64') # int -> float
print(A_float1, A_float1.dtype)

A_int2 = A_float1.astype('int32') # float -> int 
print(A_int2, A_int2.dtype)

A_float2 = np.array([1.1, 2.1, 3.1])
A_int3 = A_float2.astype('int32') # float -> int
print(A_int3, A_int3.dtype)
```

### [ ndarray axis   ndarray.sum ]

```python
A = np.array([[1,2,3],
            [2,3,4]])

print(A.sum()) #    
print(A.sum(axis=0)) #   
print(A.sum(axis=1)) #   
```

### [ ndarray   - arange, zeros, ones ]

```python
# arange 
Arange_A = np.arange(10)
print(Arange_A)
print(Arange_A.dtype, Arange_A.shape, end="\n\n")

# arange  - start, stop 
Arange_A = np.arange(start=1, stop=10)
print(Arange_A)
print(Arange_A.dtype, Arange_A.shape, end="\n\n")

# zeros 
Zero_A = np.zeros((3,2), dtype='int32')
print(Zero_A)
print(Zero_A.dtype, Zero_A.shape, end="\n\n")

# ones 
Ones_A = np.ones((3,2)) # dtype   float64
print(Ones_A)
print(Ones_A.dtype, Ones_A.shape)
```

### [ ndarray shape  ndarray.reshape ]

```python
# ndarray 
Ori_A = np.arange(10)
print("Ori_A:\n", Ori_A)

# (2,5) reshape
A1 = Ori_A.reshape(2,5)
print("A1:\n", A1)

# (5,2) reshape
A2 = Ori_A.reshape(5,2)
print("A2:\n", A2)
```

####    shape   

```python
Ori_A.reshape(4,3)
```

#### reshape  -1     ndarray 

```python
Ori_A = np.arange(10)
print("Ori_A:\n", Ori_A, end="\n\n")

# (-1,5) -> colums(axis=1)  5 \n  
A1 = Ori_A.reshape(-1, 5)
print("A1 shape: ", A1.shape)
print("A1:\n", A1, end="\n\n")

# (5, -1) -> row(axis=0)  5   
A2 = Ori_A.reshape(5, -1)
print("A2 shape: ", A2.shape)
print("A2:\n", A2)
```

####      reshape (-1, 1), (-1, )

```python
Ori_A = np.arange(5)

# 1 -> 2 (-1, 1) 
A_2d = Ori_A.reshape(-1, 1)
print("A_2d shape: ", A_2d.shape)
print("A_2d:\n", A_2d, end="\n\n")

# 2 -> 1 (-1, )
A_1d = Ori_A.reshape(-1, )
print("A_1d shape: ", A_1d.shape)
print("A_1d:\n", A_1d)
```

### [ (Indexing)  ndarray    ]

### -   1.     

#### index     

```python
# ndarray 
A = np.arange(10)
print("A: ", A)

# index     
Value = A[2]
print("Value: ", Value)
print(type(Value))
```

#### index -1    .

```python
print("  : ", A[-1], "   : ", A[-2])
```

#### index   ndarray update  .

```python
A[0] = 9
A[8] = 0
print("A: ", A)
```

#### 2 ndarray     

```python
# 2 ndarray 
A_1d = np.arange(9)
A_2d = A_1d.reshape(3,3)
print(A_2d)

#    
print("(row=0, col=0) index  : ", A_2d[0,0])
print("(row=1, col=0) index  : ", A_2d[1,0])
print("(row=0, col=2) index  : ", A_2d[0,2])
print("(row=2, col=2) index  : ", A_2d[2,2])
```

### -   2.  (Slicing)

#### 1 ndarray 

```python
Ori_A = np.arange(9)
A1 = Ori_A[:3]
print(A1)

A2 = Ori_A[3:]
print(A2)

A3 = Ori_A[:]
print(A3)
```

#### 2 ndarray 

```python
A_1d = np.arange(9)
A_2d = A_1d.reshape(3,3)
print("A_2d\n", A_2d, end="\n\n")

print("A_2d[0:2, 0:2]\n", A_2d[0:2, 0:2], end="\n\n")
print("A_2d[1:3, 0:3]\n", A_2d[1:3, 0:3], end="\n\n")
print("A_2d[1:3, :]\n", A_2d[1:3, :], end="\n\n")
print("A_2d[:, :]\n", A_2d[:, :], end="\n\n")
```

### -   3.   (fancy indexing)

```python
A_1d = np.arange(9)
A_2d = A_1d.reshape(3,3)
print("A_2d\n", A_2d, end="\n\n")

A1 = A_2d[[0,1], 2]
print("A_2d[[0,1],2] => ", A1.tolist())

A2 = A_2d[[0,2], 0:2]
print("A_2d[[0,2], 0:2] => ", A2.tolist())

A3 = A_2d[[0,1]]
print("A_2d[[0,1]] => ", A3.tolist())
```

### -   4.   (Boolean indexing)

####  boolean .

```python
A_1d = np.arange(9)
print("A_1d: ", A_1d)

print(A_1d > 5)

Var = A_1d > 5
print("Var: ", Var)
print(type(Var))
```

####    &   

```python
#   
print("A_1d: ", A_1d)
Boolean_A1 = A_1d[A_1d > 5]
print("A_1d > 5     : ", Boolean_A1)

#   
B_index = np.array([False, False, False, False, False, False,  True,  True,  True])
Boolean_A2 = A_1d[B_index]
print("    : ", Boolean_A2)
```

####   X

```python
A_1d = np.arange(9)
Tar = []

for i in range(0, 9):
    if A_1d[i] > 5:
        Tar.append(A_1d[i])
        
A_seleted = np.array(Tar)
print(A_seleted)
```

####   O

```python
A_1d = np.arange(9)
print(A_1d[A_1d > 5])
```

### - sort()

#### np.sort()   .    .

```python
Ori_A = np.array([3,1,9,5])
print(" : ", Ori_A)

Sort_A1 = np.sort(Ori_A)
print("  : ", Sort_A1)
print(" : ", Ori_A)
```

#### ndarray.sort()  .   .

```python
Ori_A = np.array([3,1,9,5])
print(" : ", Ori_A)

Sort_A2 = Ori_A.sort()
print("  : ", Sort_A2)
print(" : ", Ori_A)
```

####     [::-1]

```python
SortDesc_A = np.sort(Ori_A)[::-1]
print(" : ", SortDesc_A)
```

#### 2 ndarray 

```python
A_2d = np.array([[8, 12],
               [7, 1]])

Sort_A_2d0 = np.sort(A_2d, axis=0)
print("  :\n", Sort_A_2d0, end="\n\n")

Sort_A_2d1 = np.sort(A_2d, axis=1)
print("  :\n", Sort_A_2d1)
```

### -   argsort()

####   argsort()

```python
Ori_A = np.array([3,1,9,5])
print(np.sort(Ori_A))

Sort_indices = np.argsort(Ori_A)
print(type(Sort_indices))
print("     : ", Sort_indices)
```

####    argsort()

```python
Ori_A = np.array([3,1,9,5])
print(np.sort(Ori_A)[::-1])

Sort_indices = np.argsort(Ori_A)[::-1]
print(type(Sort_indices))
print("      : ", Sort_indices)
```

#### argsort  -   

```python
Name_A = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
Score_A = np.array([78, 95, 84, 98, 88])

#     
Sort_indices = np.argsort(Score_A)
print("sort indices: ", Sort_indices)

#    
Sort_Name_A = Name_A[Sort_indices]
Sort_Score_A = Score_A[Sort_indices]

print(Sort_Name_A)
print(Sort_Score_A)
```
