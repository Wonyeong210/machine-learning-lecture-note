# Pandas DataFrame 

## [ Pandas DataFrame 기본 API ]

**read_csv()을 이용하여 csv -> DataFrame을 한다.**                                                                                 
**sep 인자로 다른 분리자를 설정해서 로드할 수도 있다.**

```python
import pandas as pd

T_Df = pd.read_csv("./titanic_train.csv")  # sep="\t"로 변경 가능
print("titanic 변수 type: ", type(T_Df))
```

#### head() : DataFrame의 맨 앞 데이터 추출 가능. 기본 크기는 5이다.

```python
T_Df.head()
```

```python
T_Df.head(3)
```

**dictionary로 DataFrame 생성, 새로운 column 추가, index에 새로운 값 할당** 

```python
# dictionary 생성
Dic = {'Name': ['Chulmin', 'Eunkyung','Jinwoong','Soobeom'],
        'Year': [2011, 2016, 2015, 2015],
        'Gender': ['Male', 'Female', 'Male', 'Male']
       }

# dictinary -> DataFrame
D_Df = pd.DataFrame(Dic)
print(D_Df)
print("#"*30 + "\n")

# 새로운 컬럼명 추가
D_Df = pd.DataFrame(Dic, columns=["Name", "Year", "Gender", "Age"])
print(D_Df)
print("#"*30 + "\n")

# 인덱스에 새로운 값 할당
D_Df = pd.DataFrame(Dic, index=["one", "two", "three", "four"])
print(D_Df)
print("*NaN은 값이 없다는 뜻이다.")
print("#"*30 + "\n")
```

**DataFrame의 column명 출력**

```python
print("column: ", T_Df.columns)
```

**DataFrame의 index 출력**

```python
print("index: ", T_Df.index)
```

**DataFrame의 index values 출력**

```python
print("index value", T_Df.index.values)
```

**DataFrame에서 일부 Series 추출 - column 한 개**

```python
S = T_Df["Name"]
print(S.head(3))
print("type: ", type(S))
```

**DataFrame에서 일부 DataFrame 추출 - column 한 개**

```python
OneCol_Df = T_Df[["Name"]]
print(OneCol_Df.head(3))
print("type: ", type(OneCol_Df))
```

**DataFrame에서 일부 DataFrame 추출 - column 여러개**

```python
F_Df = T_Df[["Name", "Age"]]
print(F_Df.head(3))
print("type: ", type(F_Df))
```

**shape : DataFrame의 row, column 크기를 가진다.**

```python
print("DataFrame의 크기: ", T_Df.shape)
```

**info() : DataFrame내의 column명, data type, null 건수, data 건수 정보 제공**

```python
T_Df.info() # 결과 중 object 는 string으로 이해
```

**describe() : 숫자형 column들에 대해서 데이터 값의 평균, 표준편차, 4분위 분포도를 제공.**

```python
T_Df.describe()
```

**value_counts() : 동일한 개별 데이터 값이 몇건 있는 지 정보 제공. 즉 개별 데이터 값의 분포도를 제공 (단, Series 객체에서만 호출 가능)**

```python
V_Counts = T_Df["Pclass"].value_counts()
print(V_Counts)
```

**sort_values() : by=정렬컬럼, ascending=True/False는 오름차순/내림차순**

#### *하나의 컬럼으로 정렬*

```python
T_Df.sort_values(by="Pclass", ascending=True).head()
```

#### *컬럼으로 DataFrame 일부 추출 후 정렬*

```python
T_Df[["Name", "Age"]].sort_values(by="Age").head()
```

#### *여러개의 컬럼으로 정렬*

```python
T_Df[["Name", "Age", "Pclass"]].sort_values(by=["Pclass", "Age"]).head()
```
