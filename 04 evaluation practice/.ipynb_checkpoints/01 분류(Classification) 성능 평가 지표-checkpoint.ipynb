## [ Accuracy(정확도) ]

* 직관적으로 모델 예측 성능을 나타내는 평가 지표.
* 이진 분류의 경우, 데이터의 구성에 따라 머신러닝 성능이 왜곡 가능하여 정확도만으로 성능을 평가하지 않음.

### 1) 타이타닉 생존 예측 정확도

#### predict 메소드를 단순히 성별로 사망, 생존을 결정하도록 변경

```python
import numpy as np
from sklearn.base import BaseEstimator

class MyDummyClassifier(BaseEstimator): # BaseEstimator를 상속하는 class 생성
    # fit() 메소드가 아무것도 학습하지 않게 함.
    def fit(self, X, y=None):
        pass
    
    # predict() 메소드는 단순히 "Sex" feature가 1이면 0, 그렇지 않으면 1로 예측
    # 즉, 여자면 생존, 남자면 사망으로 예측
    def predict(self, X):
        pred = np.zeros((X.shape[0], 1))
        for i in range (X.shape[0]):
            if X["Sex"].iloc[i] == 1:
                pred[i] = 0
            else:
                pred[i] = 1
        return pred
```

#### 타이타닉 데이터 전처리 함수

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Null 처리 함수
def fillna(df) :
    df["Age"].fillna(df["Age"].mean(), inplace=True)
    df["Cabin"].fillna("N", inplace=True)
    df["Embarked"].fillna("N", inplace=True)
    df["Fare"].fillna(0, inplace=True)
    return df

# 머신러닝 알고리즘에 불필요한 속성 제거
def drop_features(df):
    df.drop(["PassengerId", "Name", "Ticket"], axis=1, inplace=True)
    return df

# 레이블 인코딩 수행
def format_features(df):
    df["Cabin"] = df["Cabin"].str[:1]
    features = ["Cabin", "Sex", "Embarked"]
    for feature in features:
        le = LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    return df

# 앞에서 생성한 함수들 호출
def transform_features(df):
    df = fillna(df)
    df = drop_features(df)
    df = format_features(df)
    return df
```

#### 타이타닉 데이터 로딩, 가공, 분할

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 원본 데이터 로딩
titanic_df = pd.read_csv("./titanic_train.csv")

# 원본 데이터 가공, 전처리
y_titanic_df = titanic_df["Survived"]
X_titanic_df = titanic_df.drop("Survived", axis=1)
X_titanic_df = transform_features(X_titanic_df)

# 학습/테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_titanic_df, y_titanic_df,
                                                   test_size=0.2, random_state=0)
```

#### 위에서 생성한 Dummy Classifier로 학습/예측/평가 수행

```python
myclf = MyDummyClassifier()
myclf.fit(X_train, y_train)
mypredictions = myclf.predict(X_test)

print("Dummy Classifier의 정확도는: {0: .4f}".format(accuracy_score(y_test, mypredictions)))
```

 ### 2. MNIST 데이터 예측 정확도

#### MNIST 데이터 로딩

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd

digits = load_digits()

print(digits.data)
print("### digits.data.shape", digits.data.shape, end="\n\n")

print(digits.target)
print("### digits.target.shape", digits.target.shape)
```

#### 모두 0 (false)값으로 만들어서 반환하는 MyFakeClassifier 클래스 생성

```python
class MyFakeClassifier(BaseEstimator): # BaseEstimator를 상속받는 MyFakeClassifier 클래스
    def fit(self, X, y):
        pass
    
    # 입력으로 들어오는 X 데이터 세트의 크기만큼 모두 0값으로 만들어서 반환
    def predict(self, X):
        return np.zeros((len(X), 1), dtype=bool)
```

#### digits번호가 7이면 True, 아니면 False인데 이를 astype(int)로 각각 1, 0으로 변환 

```python
print(digits.target == 7)
y = (digits.target==7).astype(int)
```

#### 학습/테스트 데이터 분할

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, y, random_state=11)
```

#### 불균형한 Label 데이터 분포도 확인

```python
print("레이블 테스트 세트 크기: ", y_test.shape)
print("테스트 세트 레이블 0과 1의 분포도")
print(pd.Series(y_test).value_counts())
```

#### MyFakeClassifier로 학습/예측/정확도 평가

```python
fakeclf = MyFakeClassifier()
fakeclf.fit(X_train, y_train)
fakepred = fakeclf.predict(X_test)
print("모든 예측을 0으로 해도 정확도는: {: .3f}".format(accuracy_score(y_test, fakepred)))
```

## [ 정밀도(Precison) & 재현율(Recall) ]

#### 정밀도
* 정밀도 = TP / (FP + TP)
* 정밀도는 "예측"을 Positive로 한 대상 중에 예측과 실제 값이 Positive로 일치한 데이터의 비율을 뜻한다.
* precision_score()
* 정밀도가 상대적으로 더 중요한 경우: 실제 Negative인 데이터 예측을 Positive로 잘못 판단하게 되면 업무상 큰 영향이 발생하는 경우 (ex. 스팸 메일)

#### 재현율
* 재현율 = TP / (FN + TP)
* 재현율은 "실제 값"이 Positive로 한 대상 중에 예측과 실제 값이 Positive로 일치한 데이터의 비율을 뜻한다.
* recall_score()
* 재현율이 상대적으로 더 중요한 경우: 실제 Positive인 데이터 예측을 Negative로 잘못 판단하게 되면 업무상 큰 영향이 발생하는 경우 (ex. 암 진단, 금융사기 판별)

#### MyFakeClassifier의 예측 결과로 정밀도 & 재현율 측정

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

print("정밀도: ", precision_score(y_test, fakepred))
print("재현율: ", recall_score(y_test, fakepred))
```

#### 앞으로 자주 쓸 함수 get_clf_eval : 오차행렬, 정확도, 정밀도, 재현율을 한꺼번에 계산하는 함수

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

def get_clf_eval(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    
    print("오차 행렬")
    print(confusion)
    print("정확도: {0: .4f}, 정밀도: {1: .4f}, 재현율: {2: .4f}".format(accuracy, precision, recall))
```

#### 실습) 타이타닉 데이터를 예측하고 오차행렬, 정확도, 정밀도, 재현율을 계산

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# 원본 데이터 로딩
titanic_df = pd.read_csv("./titanic_train.csv")

# 데이터 전처리
y_titanic_df = titanic_df["Survived"]
X_titanic_df = titanic_df.drop("Survived", axis=1)
X_titanic_df = transform_features(X_titanic_df)

# 학습/테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_titanic_df, y_titanic_df, 
                                                   test_size=0.2, random_state=11)

# LogisticRegression() 객체 생성
lr_clf = LogisticRegression()

# 예측
lr_clf.fit(X_train, y_train)
pred = lr_clf.predict(X_test)

# 오차행렬, 정확도, 정밀도, 재현율 계산
get_clf_eval(y_test, pred)
```

## [ 정밀도/재현율 Trade-off ]

* 업무 특성상 정밀도or재현율을 조정해야할 경우 분류의 결정 임계값(Threshold)를 조정해 정밀도or재현율을 높일 수 있음.
* 정밀도/재현율의 Trade-off : 정밀도와 재현율을 상호 보완적이므로 어느 한쪽을 높이면 다른 하나는 떨어지기 쉬움
* 분류 결정 임계값(Threshold)이 낮아질 수록 Positive로 예측할 확률이 높아짐. 즉, 재현율 증가
* predict_proba() 메소드 : 분류 결정 예측 확률을 반환
* precision_recall_curve() 함수 : 임계값에 따른 정밀도, 재현율의 변화값 제공

#### predict_proba() 메소드 : 분류 결정 예측 확률 ([0일 확률, 1일 확률], ....) 을 반환

```python
pred_proba = lr_clf.predict_proba(X_test)
pred = lr_clf.predict(X_test)

# pred_proba() 결과 출력
print("pred_proba()결과 shape :{0}".format(pred_proba.shape))
print("pred_proba array에서 앞 3개만 샘플로 추출: \n", pred_proba[:3])

# (예측 확률 array / 예측 결과값) 같이 한 눈에 출력
pred_proba_result = np.concatenate([pred_proba, pred.reshape(-1, 1)], axis=1)
print("\n두 개의 class 중에서 더 큰 확률을 클래스 값으로 예측\n", pred_proba_result[:3])
```

#### 임계값에 따른 정밀도, 재현율의 변화값 제공하는 precision_recall_curve() 활용

```python
from sklearn.metrics import precision_recall_curve

# label 값이 1일 때의 예측 확률을 추출
pred_proba_class1 = lr_clf.predict_proba(X_test)[:, 1]

# 실제값 데이터 세트와 label 값이 1일 떄의 예측 확률을 precision_recall_curve 인자로 입력
precisions, recalls, thresholds = precision_recall_curve(y_test, pred_proba_class1)
print("반환된 thresholds 배열의 shape: ", thresholds.shape)
print("반환된 precisions 배열의 shape: ", precisions.shape)
print("반환된 recalls 배열의 shape: ", recalls.shape)

print("\nthresholds 5 sample:", thresholds[:5])
print("precisions 5 sample:", precisions[:5])
print("recalls 5 sample:", recalls[:5])

# 반환된 thresholds 배열 row가 147건 이므로 샘플로 10건만 추출.
# threshold는 15 step으로 추출.
thr_index = np.arange(0, thresholds.shape[0], 15)
print("\n샘플 추출을 위한 threshold 배열의 index 10개: ", thr_index)
print("샘플용 10개의 threshold: ", np.round(thresholds[thr_index], 2))

# 15 step 단위로 추출된 threshold에 따른 정밀도, 재현율
print("\n샘플 threshold에 따른 정밀도: ", np.round(precisions[thr_index], 3))
print("샘플 threshold에 따른 재현율: ", np.round(recalls[thr_index],3))
```

#### threshold 변경에 따른 정밀도, 재현율 변화 곡선을 시각화

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline

def precision_recall_curve_plot(y_test, pred_proba_c1):
    # threshold ndarray와 이 threshold에 따른 정밀도, 재현율 ndarray를 추출
    precisions, recalls, thresholds = precision_recall_curve(y_test, pred_proba_c1)
    
    # X축(threshold), Y축(정밀도/재현율) 각각 plot 수행. 정밀도는 점선
    plt.figure(figsize=(8, 6))
    threshold_boundary = thresholds.shape[0]
    plt.plot(thresholds, precisions[0:threshold_boundary], linestyle="--", label="precision")
    plt.plot(thresholds, recalls[0:threshold_boundary], label="recall")
    
    # threshold 값 X축의 Scale을 0.1 단위로 변경
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1), 2))
    
    # x축, y축의 labe, legend, grid 설정
    plt.xlabel("Threshold value")
    plt.ylabel("Precision and Recall value")
    plt.legend()
    plt.grid()
    plt.show()

# 함수 호출
precision_recall_curve_plot(y_test, lr_clf.predict_proba(X_test)[:, 1])
```

## [ ROC Curve & AUC ]

#### ROC 곡선
* ROC 곡선 (Receiver Operation Characteristic Curve)
* FPR이 변할 때 TPR가 어떻게 변하는 지를 나타내는 곡선임.
* X축(FPR), Y축(TPR)로 잡으면 FPR의 변화에 따른 TPR의 변화가 곡선 형태로 나타남.
* roc_curve(y_true, y_score)

#### AUC
* AUC (Area Under Curve)
* 분류의 성능 지표로 사용됨.
* ROC 곡선 밑의 면적을 구한 것임. 일반적으로 1에 가까울 수록 좋은 수치임.
* roc_auc_score(y_ture, y_score)

#### TPR
* TPR (True Positive Rate)
* 재현율(민감도)을 나타냄
* TPR = TP / (FN + TP)

#### FPR
* 실제 Negative을 잘못 예측한 비율
* 즉, 실제는 Negative인데 Positive or Negative로 예측한 것 중 Positive로 잘못 예측한 비율임.
* FPR = FP / (FP + TN)

    
 
* 정밀도 = TP / (FP + TP)
* 재현율 = TP / (FN + TP)
